```python
2. CHECK PALINDROME OR NOT
```

```python
def is_palindrome(s):
  """checks if a string is a palindrome."""
  return s == s[::-1]
word1 = "racecar"
word2 = "hello"
print(f"'{word1}' is a palindrome: {is_palindrome(word1)}")
print(f"'{word2}' is a palindrome: {is_palindrome(word2)}")
```

```python
3. CHECK PRIME OR NOT PRIME
```

```python
num = int(input("enter a number: ")) 
if num <= 1:
    print(f"{num} ,is not a prime number")
else:
    for i in range(2,num):
        if num % i == 0:
            print(f"{num},is not a prime number")
            break
    else:
        print(f"{num},is a prime number")
    
```

```python
15. CHECK ARMSTONG NUMBER
```

```python
num = int(input("enter the number: "))
power = len(str(num))
temp = num
total = 0
while temp >0:
    digit = temp%10
    total += digit**power
    temp //= 10
if total == num:
    print("{num} is an armstrong number")
else:
    print("{num} is  not an armstrong number")
    
```

```python
21. FIND COMMON ELEMENTS BETWEEN TWO LISTS
```

```python
list1 = [1,2,3,4,6,8,9]
list2 = [2,1,4,3,67,435]
common = []
for item1 in list1:
    for item2 in list2:
        if item1 == item2:
            if item1 not in common:
                common.append(item1)
print("common elements:", common)
```

```python
8.CREATE VOWELS AND CONSONANTS IN A STRING
```

```python
text = (input("enter a string: "))
vowels = "aeiou"
vowels_count = 0
consonant_count = 0
for i in text:
    if i.isalpha():
        if i in vowels:
            vowels_count +=1
        else:
            consonant_count += 1
print("vowels:",vowels_count)
print("consonants:", consonant_count)

            
```

```python
7. SORT IN ASCENDING AND DECENDING ORDER
```

```python
numbers= [34,6,9,86,54,52]
n = len(numbers)
for i in range(n):
    for j in range(0, n - i - 1):
        if numbers[j]< numbers[j+1]:
            numbers[j], numbers[j+1] = numbers[j+1],numbers[j]
print("decending:", numbers)
```
